```python
from dataset.cell_motion import CellMotionRollout
from model.simulator import Simulator
import torch
from torch_geometric.loader import DataLoader
from torch_geometric.data import Data
import torch_geometric.transforms as T
import numpy as np
import os
from typing import Optional
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from scipy.spatial import cKDTree, Delaunay

# 设置计算设备
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# 设置路径
load_dir = 'checkpoint/20260204_161356/best_epoch/best_checkpoint.pth'
dataset_dir = os.path.join(os.getcwd(), 'dataset')

# 训练/推理参数
batch_size = 1
sequence_length = 20  # 历史帧数
max_rollout_steps = 40  # 默认 rollout 步数
time_step = 1.0  # 时间步长（与训练时一致）
correction_interval = 15
# 网络参数（应与训练时一致）
hidden_size = 256
message_passing_num = 15
node_input_size = sequence_length * 2 + 1  # e.g. 20帧 * 2维 + 1时间
edge_input_size = 3
learning_rate = 1.811923871799656e-04

# 创建模拟器并加载 checkpoint
simulator = Simulator(
    message_passing_num=message_passing_num,
    node_input_size=node_input_size,
    edge_input_size=edge_input_size,
    device=device,
    hidden_size=hidden_size,
    model_dir=load_dir
)

try:
    loaded = simulator.load_checkpoint()  # 期望它设置 simulator.model 并返回模型（或 None）
except TypeError:
    # 如果 load_checkpoint 需要参数（例如路径），使用给定路径
    loaded = simulator.load_checkpoint(load_dir)

# 确保模型存在并处于 eval 模式
if hasattr(simulator, 'model') and simulator.model is not None:
    model = simulator.model
else:
    # 如果 load_checkpoint 返回模型对象则用它
    model = loaded
if model is None:
    raise RuntimeError("无法加载模型：请检查 Simulator.load_checkpoint API 与提供的路径。")
model.to(device)
model.eval()


# transforms（计算边特征）
transformer = T.Compose([
    T.Cartesian(norm=False),
    T.Distance(norm=False)
])
def build_edges_from_geometry(pos, original_edge_attr=None, method='knn', k=6, 
                               radius=None, self_loop=False, device='cpu'):
    """
    基于节点几何位置构建邻接关系，并生成与原始维度匹配的边属性
    
    Args:
        pos: (N, 2) 节点位置坐标 (torch.Tensor 或 np.ndarray)
        original_edge_attr: 原始边属性 (用于获取维度信息)
        method: 邻接构建方式，可选 'knn', 'radius', 'delaunay'
        k: KNN 方法中的邻域数量
        radius: 半径方法中的距离阈值
        self_loop: 是否包含自环
        device: 张量所在设备
    
    Returns:
        edge_index: (2, E) LongTensor，边索引
        edge_attr: (E, D) FloatTensor，边属性（维度与输入一致）
    """
    
    # 确保输入是 numpy 数组
    if isinstance(pos, torch.Tensor):
        pos_np = pos.cpu().detach().numpy()
    else:
        pos_np = np.array(pos)
    
    # 获取原始边属性的维度
    if original_edge_attr is not None:
        if isinstance(original_edge_attr, torch.Tensor):
            edge_attr_dim = original_edge_attr.shape[1] if original_edge_attr.dim() > 1 else 1
        else:
            edge_attr_dim = original_edge_attr. shape[1] if len(original_edge_attr. shape) > 1 else 1
    else:
        edge_attr_dim = 2  # 默认维度：距离 + 幅值
    
    N = pos_np.shape[0]
    
    if method == 'knn':
        edge_index, distances = _build_knn_edges(pos_np, k, self_loop)
    elif method == 'radius':
        if radius is None:
            radius = np.mean(np.linalg.norm(pos_np - pos_np. mean(axis=0), axis=1)) * 0.3
        edge_index, distances = _build_radius_edges(pos_np, radius, self_loop)
    elif method == 'delaunay':
        edge_index, distances = _build_delaunay_edges(pos_np, self_loop)
    else:
        raise ValueError(f"未知的构建方法: {method}")
    
    # 根据距离生成边属性，匹配维度
    if edge_index.shape[1] > 0:
        edge_attr = _generate_edge_attributes(
            pos_np, edge_index, distances, edge_attr_dim
        )
    else:
        edge_attr = np.zeros((0, edge_attr_dim), dtype=np.float32)
    
    # 转换为 torch 张量
    edge_index = torch.tensor(edge_index, dtype=torch.long, device=device)
    edge_attr = torch.tensor(edge_attr, dtype=torch.float32, device=device)
    
    return edge_index, edge_attr


def _build_knn_edges(pos_np, k, self_loop=False):
    """KNN 方法构建边"""
    N = pos_np.shape[0]
    k = min(k, N - 1)
    
    tree = cKDTree(pos_np)
    distances, indices = tree.query(pos_np, k=k + 1)
    
    edge_index_list = []
    distances_list = []
    
    for i in range(N):
        neighbors = indices[i, 1:]
        dists = distances[i, 1:]
        
        for j, neighbor in enumerate(neighbors):
            if neighbor < N:
                edge_index_list.append([i, neighbor])
                distances_list.append(dists[j])
    
    edge_index = np.array(edge_index_list). T if edge_index_list else np. zeros((2, 0), dtype=np.long)
    distances = np.array(distances_list)
    
    return edge_index, distances


def _build_delaunay_edges(pos_np, self_loop=False):
    """基于 Delaunay 三角剖分构建边"""
    N = pos_np.shape[0]
    
    if N < 3:
        return _build_knn_edges(pos_np, k=min(2, N-1), self_loop=self_loop)
    
    try:
        tri = Delaunay(pos_np)
    except Exception as e:
        return _build_knn_edges(pos_np, k=min(6, N-1), self_loop=self_loop)
    
    edge_set = set()
    for simplex in tri.simplices:
        for i in range(3):
            for j in range(i + 1, 3):
                edge = tuple(sorted([simplex[i], simplex[j]]))
                edge_set.add(edge)
    
    edge_index_list = []
    distances_list = []
    
    for i, j in edge_set:
        for src, tgt in [(i, j), (j, i)]:
            dist = np.linalg.norm(pos_np[tgt] - pos_np[src])
            edge_index_list.append([src, tgt])
            distances_list.append(dist)
    
    edge_index = np.array(edge_index_list).T if edge_index_list else np.zeros((2, 0), dtype=np.long)
    distances = np.array(distances_list)
    
    return edge_index, distances


def _generate_edge_attributes(pos_np, edge_index, distances, dim):
    """
    根据几何信息生成指定维度的边属性
    
    Args:
        pos_np: (N, 2) 节点位置
        edge_index: (2, E) 边索引
        distances: (E,) 距离数组
        dim: 目标维度
    
    Returns:
        edge_attr: (E, dim) 边属性
    """
    E = edge_index.shape[1]
    edge_attr = np.zeros((E, dim), dtype=np.float32)
    
    if E == 0:
        return edge_attr
    
    # 第 0 列：距离
    edge_attr[:, 0] = distances
    
    if dim >= 2:
        # 第 1 列：方向向量的 x 分量
        directions = pos_np[edge_index[1]] - pos_np[edge_index[0]]
        edge_attr[:, 1] = directions[:, 0]
    
    if dim >= 3:
        # 第 2 列：方向向量的 y 分量
        edge_attr[:, 2] = directions[:, 1]
    
    if dim >= 4:
        # 第 3 列：距离的平方
        edge_attr[:, 3] = distances ** 2
    
    if dim >= 5:
        # 第 4 列：向量的幅值（应该等于距离，作为冗余特征）
        edge_attr[:, 4] = np.linalg.norm(directions, axis=1)
    
    # 如果维度更高，填充相同的特征或零
    if dim > 5:
        edge_attr[:, 5:] = 0.0
        
    return edge_attr
def compute_velocity_features(graph_x, sequence_len):
    """
    辅助函数：将绝对位置特征转换为速度（差分）特征。
    假设 graph.x 的结构是 [batch_size, sequence_length * 2 + 1]
    其中前 sequence_length * 2 是 (x_t-k, y_t-k, ..., x_t, y_t)
    
    如果不修改 Dataset 类，我们需要在这里实时计算差分。
    """
    # 剥离时间特征（假设它是最后一个维度）
    time_feature = graph_x[:, -1:]
    pos_features = graph_x[:, :-1]
    
    # 重塑为 [num_nodes, sequence_length, 2]
    # 注意：这里需要根据您 Dataset 中具体的 flatten 顺序来调整
    num_nodes = pos_features.size(0)
    pos_seq = pos_features.view(num_nodes, sequence_len, 2)
    
    # 计算速度 v_t = p_t - p_{t-1}
    # 对于第一个时间步，由于没有前一个时间步，我们可以补0或者复制第二个时间步的速度
    # 这里采用计算相邻差分：结果形状 [num_nodes, sequence_len - 1, 2]
    velocities = pos_seq[:, 1:, :] - pos_seq[:, :-1, :]
    
    # 为了保持输入维度一致（或者您可以根据需要填充），通常需要处理掉减少的一个时间步
    # 方法1：填充第一帧速度为0
    zeros = torch.zeros(num_nodes, 1, 2, device=graph_x.device)
    velocities_padded = torch.cat([zeros, velocities], dim=1)
    
    # 展平回 [num_nodes, sequence_length * 2]
    velocities_flat = velocities_padded.view(num_nodes, -1)
    
    # 拼回时间特征
    new_features = torch.cat([velocities_flat, time_feature], dim=1)
    
    return new_features

def inference(num_rollout_steps: int = 10, edge_method='knn', edge_k=6, 
              start_frame_idx: int = 0, correction_interval: int = 3):
    """
    Args:
        num_rollout_steps: 预测步数
        edge_method: 建边方法
        edge_k: KNN的k值
        start_frame_idx: 从第几帧开始预测
        correction_interval: 每隔多少步用GT校正一次
    """
    
    all_predictions = []
    all_gt_trajectories = []  
    all_initial_positions = []
    all_histories = []
    all_correction_flags = []

    # 历史特征长度：Sequence Length * 2 (x,y)
    num_history_features = sequence_length * 2 
    
    # ===== 第一阶段：收集所有帧的batch数据 =====
    all_frame_batches_raw = []
    
    # 注意：这里假设 test_loader 顺序返回时间序列数据
    for batch_idx, batch_graph in enumerate(test_loader):
        try:
            batch_graph = transformer(batch_graph)
        except Exception:
            batch_graph = transformer(batch_graph.to(device)).to('cpu')
        
        batch_graph = batch_graph.to(device)
        all_frame_batches_raw.append(batch_graph)
    
    if start_frame_idx >= len(all_frame_batches_raw):
        raise ValueError(f"start_frame_idx ({start_frame_idx}) 不能大于或等于总帧数")
        
    all_frame_batches = all_frame_batches_raw[start_frame_idx:]
    num_frames = len(all_frame_batches)
    first_batch = all_frame_batches[0]
    num_graphs_per_batch = len(first_batch.ptr) - 1
    
    effective_rollout_steps = min(num_rollout_steps, num_frames - 1)
    
    print(f"第二阶段：执行Rollout预测")
    
    for graph_idx in range(num_graphs_per_batch):
        
        initial_frame_batch = all_frame_batches[0]
        start_node = initial_frame_batch.ptr[graph_idx].item()
        end_node = initial_frame_batch.ptr[graph_idx + 1].item()
        node_indices = torch.arange(start_node, end_node)
        
        # 初始位置
        initial_positions = initial_frame_batch.pos[node_indices].clone()
        all_initial_positions.append(initial_positions.cpu().detach().numpy())

        # 初始化当前状态
        current_x = initial_frame_batch.x[node_indices].clone()
        current_pos = initial_positions.clone()
        
        # 记录初始历史用于可视化
        history_flat = current_x[:, :num_history_features]
        history_unflattened = history_flat.view(history_flat.shape[0], sequence_length, 2)
        history_trajectory = history_unflattened.permute(1, 0, 2).cpu().detach().numpy()
        current_pos_np = initial_positions.cpu().detach().numpy()
        current_pos_expanded = np.expand_dims(current_pos_np, axis=0)
        full_history = np.concatenate([history_trajectory, current_pos_expanded], axis=0)
        all_histories.append(full_history)

        # 结果容器
        sample_predictions = [initial_positions.cpu().detach().numpy()]
        sample_correction_flags = [False]
        sample_gt_trajectory = [initial_positions.cpu().detach().numpy()]
        
        # 预先获取所有GT位置和GT特征（关键修正点）
        gt_data_all = [] # 存储每一步的完整batch data，用于提取正确的历史特征
        gt_positions_all = [initial_positions.cpu().detach().numpy()]
        
        # 注意：这里需要收集足够的数据
        for frame_idx in range(effective_rollout_steps + 1):
            if frame_idx < num_frames:
                frame_batch = all_frame_batches[frame_idx]
                
                # 提取特定graph的数据
                s_n = frame_batch.ptr[graph_idx].item()
                e_n = frame_batch.ptr[graph_idx + 1].item()
                indices = torch.arange(s_n, e_n)
                
                pos = frame_batch.pos[indices]
                x_feat = frame_batch.x[indices] # 包含正确的历史轨迹
                
                if frame_idx > 0: # 0帧已经在上面初始化了
                    sample_gt_trajectory.append(pos.cpu().detach().numpy())
                    gt_positions_all.append(pos.cpu().detach().numpy())
                
                gt_data_all.append({'pos': pos, 'x': x_feat})

        # 准备初始边属性
        original_edge_attr = initial_frame_batch.edge_attr
        
        # 构建初始边
        current_edge_index, current_edge_attr = build_edges_from_geometry(
            pos=current_pos,
            original_edge_attr=original_edge_attr,
            method=edge_method,
            k=edge_k,
            self_loop=False,
            device=device
        )

        steps_since_correction = 0 

        # Rollout循环
        for step in range(effective_rollout_steps):
            
            # 判断下一帧是否校正
            # 逻辑：如果已经自由跑了correction_interval步，下一步强制归位
            use_correction = (steps_since_correction >= correction_interval) and \
                           (step + 1 < len(gt_data_all))
            
            if use_correction:
                # ========== 使用Ground Truth完全重置状态 ==========
                # 关键修复：不仅重置位置，还要重置历史特征(current_x)
                # 否则 velocity 计算会包含 (Prediction_t-1 -> GT_t) 这种巨大的跳变
                
                gt_step_data = gt_data_all[step + 1] # 获取下一时刻的真实数据
                
                # 1. 强制覆盖位置
                predicted_positions = gt_step_data['pos'].clone() 
                
                # 2. 强制覆盖历史特征（这是消除抖动的关键！）
                # 如果我们不覆盖 x，模型下一帧还会以为刚经历了一个巨大的跳跃
                current_x = gt_step_data['x'].clone()
                
                # 记录
                sample_predictions.append(predicted_positions.cpu().detach().numpy())
                sample_correction_flags.append(True)
                
                # 更新当前位置指针
                current_pos = predicted_positions
                
                # 重置计数器
                steps_since_correction = 0
                # print(f"  Graph {graph_idx}, Step {step + 1}: 使用GT完全重置 ✓")
                
            else:
                # ========== 正常模型预测 ==========
                with torch.no_grad():
                    # 注意：compute_velocity_features 会计算 x_t - x_{t-1}
                    # 如果上一步是 correction，这里的 current_x 已经是 clean 的 GT history，
                    # 所以速度计算是平滑的 (GT_{t-1} -> GT_t)，而不是跳变的。
                    
                    temp_graph = Data(
                        x=compute_velocity_features(current_x, sequence_length),
                        pos=current_pos,
                        edge_index=current_edge_index,
                        edge_attr=current_edge_attr
                    )
                    
                    pred_disp_norm = model(temp_graph)

                    if isinstance(pred_disp_norm, (tuple, list)):
                        pred_disp_norm = pred_disp_norm[0]
                    if pred_disp_norm.shape != current_pos.shape:
                        pred_disp_norm = pred_disp_norm.view_as(current_pos)

                    # 反归一化
                    pred_disp = simulator._output_normalizer.inverse(pred_disp_norm)
                
                # 更新位置
                predicted_positions = current_pos + pred_disp
                sample_predictions.append(predicted_positions.cpu().detach().numpy())
                sample_correction_flags.append(False)

                # 手动更新历史特征 (Sliding Window)
                # 只有在非 Correction 步骤才需要手动 shift
                # Correction 步骤直接读取了 GT 的 x，不需要这一步
                
                history = current_x[:, :num_history_features] # [N, seq_len * 2]
                history_shifted = history[:, 2:] # 丢弃最老的一帧 (x, y)
                new_frame = predicted_positions # 添加最新预测帧
                updated_history = torch.cat([history_shifted, new_frame], dim=1)

                time_col = current_x[:, num_history_features:num_history_features+1]
                new_time_col = time_col + time_step

                current_x = torch.cat([updated_history, new_time_col], dim=1)
                current_pos = predicted_positions
                
                steps_since_correction += 1

            # 无论是否 Correction，都需要重新建边
            # 只有当位置发生显著变化时，才需要重新建边，但为了准确性通常每步都建
            current_edge_index, current_edge_attr = build_edges_from_geometry(
                pos=current_pos,
                original_edge_attr=original_edge_attr,
                method=edge_method,
                k=edge_k,
                self_loop=False,
                device=device
            )
        
        all_predictions.append(np.stack(sample_predictions, axis=0))
        all_gt_trajectories.append(np.stack(sample_gt_trajectory, axis=0))
        all_correction_flags.append(sample_correction_flags)
        
        # print(f"Graph {graph_idx} finished.")
    
    return all_initial_positions, all_predictions, all_gt_trajectories, all_histories, all_correction_flags
def visualize_rollout_animation(
    initial_pos_list,
    predictions_list,
    gt_trajectory_list=None,
    batch_idx=0,
    interval=100,
    save_path=None,
    max_display=296,
    num_tracked_particles=5,
):
    """
    Visualize rollout results animation with frame-by-frame GT comparison,
    and plot error vs step (several particles + mean error).

    Args:
        initial_pos_list: list of (N,2) arrays - initial positions
        predictions_list: list of arrays, each shape (num_steps, N, 2) - predicted trajectories
        gt_trajectory_list: list of arrays, each shape (num_steps, N, 2) - GT trajectories
        batch_idx:  batch index to display
        interval: frame interval in milliseconds
        save_path:  save path for gif if specified
        max_display: max number of particles to display
        num_tracked_particles: how many individual particles to track in error plot
    """
    if batch_idx >= len(predictions_list):
        raise IndexError(f"batch_idx {batch_idx} exceeds list length")

    pred_trajectory = predictions_list[batch_idx]  # (T, N, 2)

    if pred_trajectory.ndim != 3:
        raise ValueError(f"Predicted trajectory should be 3D, got {pred_trajectory.ndim}D")

    num_steps, num_particles, _ = pred_trajectory.shape

    # -------- GT 处理 --------
    has_gt = False
    gt_trajectory = None

    if gt_trajectory_list is not None and batch_idx < len(gt_trajectory_list):
        gt_data = gt_trajectory_list[batch_idx]

        if gt_data is not None:
            gt_trajectory = np.array(gt_data)
            has_gt = True

            if gt_trajectory.ndim != 3:
                raise ValueError(f"GT trajectory should be 3D, got {gt_trajectory. ndim}D")

            if gt_trajectory.shape[0] != num_steps:
                raise ValueError(
                    f"GT trajectory steps {gt_trajectory.shape[0]} mismatch with prediction {num_steps}"
                )

    display_n = min(num_particles, max_display)

    # -------- 误差统计（只有有 GT 才算）--------
    if has_gt: 
        # (T, display_n, 2)
        disp_all = pred_trajectory[:, :display_n, : ] - gt_trajectory[:, :display_n, :]
        # (T, display_n)
        per_particle_error = np.linalg.norm(disp_all, axis=-1)
        # (T,)
        mean_error = per_particle_error.mean(axis=1)

        tracked_n = min(num_tracked_particles, display_n)
        tracked_indices = np.arange(tracked_n)
    else:
        per_particle_error = None
        mean_error = None
        tracked_indices = []

    # -------- 轨迹动画图像 --------
    fig, ax_main = plt.subplots(1, 1, figsize=(14, 10))

    # 坐标范围
    all_x = pred_trajectory[:, :display_n, 0]. flatten()
    all_y = pred_trajectory[:, :display_n, 1].flatten()

    if has_gt:
        all_x = np.concatenate([all_x, gt_trajectory[:, :display_n, 0].flatten()])
        all_y = np. concatenate([all_y, gt_trajectory[:, :display_n, 1].flatten()])

    if all_x.size == 0 or all_y.size == 0:
        raise RuntimeError("No data to visualize.")

    margin = 0.1 * max(
        all_x.max() - all_x.min(),
        all_y.max() - all_y.min(),
        1.0
    )

    # 主图设置
    ax_main.set_xlim(all_x.min() - margin, all_x.max() + margin)
    ax_main.set_ylim(all_y. min() - margin, all_y.max() + margin)
    ax_main.set_xlabel('X Position', fontsize=12)
    ax_main.set_ylabel('Y Position', fontsize=12)
    ax_main.set_aspect('equal')
    title = f'Predicted Trajectory - Batch {batch_idx}'
    if has_gt:
        title += ' (Blue:  Predicted, Red: Ground Truth)'
    ax_main.set_title(title, fontsize=14)

    # 预测粒子 + 轨迹
    pred_particles = ax_main.scatter(
        [], [], c='blue', s=60, alpha=0.8,
        label='Predicted', edgecolors='navy', linewidths=0.8
    )
    pred_trails = [
        ax_main.plot([], [], 'b-', alpha=0.3, linewidth=1.2)[0]
        for _ in range(display_n)
    ]

    # GT 粒子 + 轨迹
    if has_gt:
        gt_particles = ax_main.scatter(
            [], [], c='red', s=60, alpha=0.8,
            label='Ground Truth', marker='o',
            edgecolors='darkred', linewidths=0.8
        )
        gt_trails = [
            ax_main.plot([], [], 'r-', alpha=0.3, linewidth=1.2)[0]
            for _ in range(display_n)
        ]

        error_text = ax_main.text(
            0.98, 0.02, '',
            transform=ax_main.transAxes, fontsize=10,
            verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9, pad=0.8),
            family='monospace'
        )

    step_text = ax_main.text(
        0.02, 0.98, '',
        transform=ax_main.transAxes, fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.9, pad=0.8),
        family='monospace'
    )

    ax_main.legend(loc='upper right', fontsize=10, framealpha=0.95)
    ax_main.grid(True, alpha=0.3)

    # -------- 新增：误差曲线图 --------
    fig_err = None
    if has_gt: 
        fig_err, ax_err = plt.subplots(1, 1, figsize=(8, 5))

        steps = np.arange(num_steps)

        # 平均误差
        ax_err.plot(
            steps, mean_error,
            'k-', lw=2, label='Mean error'
        )

        # 若干粒子误差
        colors = plt.cm.tab10(np.linspace(0, 1, len(tracked_indices)))
        for idx, c in zip(tracked_indices, colors):
            ax_err.plot(
                steps, per_particle_error[:, idx],
                color=c, lw=1.2, alpha=0.8,
                label=f'Particle {idx}'
            )

        ax_err.set_xlabel('Step', fontsize=12)
        ax_err.set_ylabel('Displacement error (L2)', fontsize=12)
        ax_err.set_title(f'Prediction Error vs Step - Batch {batch_idx}', fontsize=14)
        ax_err.grid(True, alpha=0.3)
        ax_err.legend(fontsize=9)
        fig_err. tight_layout()

    # -------- 动画初始化 / 更新 --------
    def init():
        pred_particles.set_offsets(np.empty((0, 2)))
        for t in pred_trails:
            t.set_data([], [])
        if has_gt:
            gt_particles.set_offsets(np.empty((0, 2)))
            for t in gt_trails:
                t.set_data([], [])
            error_text.set_text('')
        step_text.set_text('')

        artists = [pred_particles] + pred_trails + [step_text]
        if has_gt:
            artists += [gt_particles] + gt_trails + [error_text]
        return artists

    def update(frame):
        # 预测位置
        pred_positions = pred_trajectory[frame, :display_n, :]
        pred_particles.set_offsets(pred_positions)

        progress = frame / max(1, (num_steps - 1))
        pred_particles.set_color(plt.cm.Blues(0.4 + 0.5 * progress))

        # 预测轨迹
        if frame > 0:
            for i in range(display_n):
                seg = pred_trajectory[: frame + 1, i, :]
                pred_trails[i].set_data(seg[:, 0], seg[:, 1])

        if has_gt:
            # GT 位置
            gt_positions = gt_trajectory[frame, : display_n, :]
            gt_particles.set_offsets(gt_positions)
            gt_particles.set_color(plt.cm.Reds(0.4 + 0.5 * progress))

            # GT 轨迹
            if frame > 0:
                for i in range(display_n):
                    seg = gt_trajectory[:frame + 1, i, :]
                    gt_trails[i].set_data(seg[:, 0], seg[:, 1])

            # 误差统计文本（当前帧）
            displacement = pred_positions - gt_positions
            mse = np.mean(displacement ** 2)
            mae = np.mean(np.abs(displacement))
            distances = np.linalg.norm(displacement, axis=1)

            error_text.set_text(
                f'Frame {frame}/{num_steps - 1}\n'
                f'─────────────────\n'
                f'MSE:        {mse:.6f}\n'
                f'MAE:       {mae:.6f}\n'
                f'Mean Dist: {np.mean(distances):.6f}\n'
                f'Max Dist:  {np.max(distances):.6f}'
            )

        step_text.set_text(
            f'Progress: {progress * 100:.1f}%\n'
            f'Particles: {display_n}'
        )

        artists = [pred_particles] + pred_trails + [step_text]
        if has_gt: 
            artists += [gt_particles] + gt_trails + [error_text]
        return artists

    anim = FuncAnimation(
        fig, update, init_func=init,
        frames=num_steps, interval=interval,
        blit=True, repeat=True
    )

    # -------- 保存动画 & 误差图 --------
    if save_path:
        try:
            save_dir = os.path.dirname(save_path)
            if save_dir and not os.path. exists(save_dir):
                os.makedirs(save_dir, exist_ok=True)

            print(f"[INFO] Saving animation to {save_path}...")
            writer = PillowWriter(fps=max(1, 1000 // interval))
            anim.save(save_path, writer=writer, dpi=100)
            print("[SUCCESS] Animation saved!")

            if has_gt and fig_err is not None:
                base, ext = os.path.splitext(save_path)
                error_fig_path = base + "_error.png"
                print(f"[INFO] Saving error figure to {error_fig_path}...")
                fig_err.savefig(error_fig_path, dpi=150)
                print("[SUCCESS] Error figure saved!")

        except Exception as e:
            print(f"[ERROR] Save failed: {e}")

    # -------- 新增：打印 GT 每帧平均运动距离 --------
    if has_gt:
        # 对所有粒子（不是 display_n，而是整个 N）统计更完整
        gt_pos = gt_trajectory  # (T, N, 2)
        gt_disp = gt_pos[1: ] - gt_pos[:-1]  # (T-1, N, 2)
        gt_step_dist = np.linalg.norm(gt_disp, axis=-1).mean(axis=1)  # (T-1,)

        print(f"[GT MOTION] Batch {batch_idx} per-frame mean displacement:")
        for t, d in enumerate(gt_step_dist, start=1):
            print(f"  Frame {t}: mean displacement = {d:.6f}")

    plt.tight_layout()
    plt.show()
    return anim


def compute_rollout_errors(predictions_list, ground_truth_list=None):
    """
    计算每个 rollout 时间步的平均预测误差（在所有 batch 和所有粒子上求平均）
    返回：numpy array, length = num_steps-1，每个元素是该时间步的 mean displacement magnitude
    """
    max_steps = max(pred.shape[0] for pred in predictions_list)
    step_errors = []  # 每个 step 收集所有 batch 的节点位移误差（或位移幅值）
    for step in range(1, max_steps):
        vals = []
        for batch_pred in predictions_list:
            if batch_pred.shape[0] <= step:
                continue
            disp = batch_pred[step] - batch_pred[step - 1]  # (N,2)
            mag = np.linalg.norm(disp, axis=1)  # (N,)
            vals.append(mag. mean())
        if len(vals) > 0:
            step_errors.append(np.mean(vals))
        else:
            step_errors. append(np.nan)
    return np.array(step_errors)


if __name__ == "__main__":
    # 加载数据集与 DataLoader
    dataset_cell = CellMotionRollout(dataset_dir=dataset_dir, sequence_length=sequence_length, split='test_e0n300')
    #train-Copy1 train test test_n800 valid valid_1
    test_loader = DataLoader(dataset=dataset_cell, batch_size=batch_size, num_workers=0)

    initial_positions, predictions, ground_truth, history_frames, correction_flags = inference(
        num_rollout_steps=max_rollout_steps, 
        edge_method='delaunay',
        start_frame_idx=200,
        correction_interval=correction_interval  
    )
    # 打印校正统计
    for i, flags in enumerate(correction_flags):
        num_pred = sum(not f for f in flags)
        num_corr = sum(flags)
        print(f"Batch {i}: 预测帧数={num_pred}, 校正帧数={num_corr}")
    

    history_window = 10
    # 裁剪GT轨迹到与预测相同的帧数
    predictions_frames = predictions[0].shape[0]  
    ground_truth_cropped = [gt[:predictions_frames] for gt in ground_truth]

    # 然后用裁剪后的GT进行可视化
    for batch_idx in range(min(3, len(predictions))):
        save_path = f'gifs/rollout_animation_with_gt_batch_{batch_idx}.gif'
        anim = visualize_rollout_animation(
            initial_positions,
            predictions,
            gt_trajectory_list=ground_truth_cropped,  # 使用裁剪后的GT ground_truth_cropped
            batch_idx=batch_idx,
            interval=200,
            save_path=save_path,
            max_display=2000,
            num_tracked_particles=0,#
        )


    # 计算并打印位移误差随时间的变化
    errors = compute_rollout_errors(predictions, ground_truth_cropped)  # 使用裁剪后的GT

    # 计算并打印位移误差随时间的变化
    errors = compute_rollout_errors(predictions, ground_truth)
    print(f"Per-step average displacement (mean over batches and particles): {errors}")
    print(f"Average displacement per step (mean of steps): {np.nanmean(errors):.6f}")
    print(f"Std of step displacements: {np.nanstd(errors):.6f}")
```

```python
from dataset.cell_motion import CellMotionRollout
from model.simulator import Simulator
import torch
from torch_geometric.loader import DataLoader
from torch_geometric.data import Data
import torch_geometric.transforms as T
import numpy as np
import os
from typing import Optional, Dict, List, Tuple
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from scipy.spatial import cKDTree, Delaunay, Voronoi
import pickle

# 设置全局字体大小
plt.rcParams.update({'font.size': 16})

# ====================== 配置参数 ======================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# 路径设置
load_dir = 'checkpoint/20260204_161356/best_epoch/best_checkpoint.pth'
dataset_dir = os.path.join(os.getcwd(), 'dataset')

# 训练/推理参数
batch_size = 1
sequence_length = 20  # 历史帧数
max_rollout_steps = 50  # 默认 rollout 步数
time_step = 1.0  # 时间步长
correction_interval = 50

# 网络参数
hidden_size = 256
message_passing_num = 15
node_input_size = sequence_length * 2 + 1
edge_input_size = 3
learning_rate = 1.811923871799656e-04
start_frame_idx = 100
# ====================== 评估指标函数 ======================

def compute_spatial_density_metrics(positions: np.ndarray, grid_size: int = 20) -> Dict:
    """计算空间密度分布指标（仅保留密度熵）"""
    from scipy.stats import entropy
    
    x, y = positions[:, 0], positions[:, 1]
    
    # 计算2D直方图
    hist, x_edges, y_edges = np.histogram2d(x, y, bins=grid_size)
    
    # 归一化为概率分布
    hist_norm = hist / (hist.sum() + 1e-10)
    
    # 计算熵
    density_entropy = entropy(hist_norm.flatten() + 1e-10)
    
    return {
        'density_entropy': density_entropy,
        'histogram': hist
    }


def compute_neighbor_statistics(positions: np.ndarray, k: int = 6) -> Dict:
    """计算邻居距离统计"""
    tree = cKDTree(positions)
    
    # 查询k+1个最近邻（包括自己）
    distances, indices = tree.query(positions, k=min(k+1, len(positions)))
    
    # 去掉自己（第一个）
    neighbor_distances = distances[:, 1:]
    
    # 统计量
    mean_nn_dist = neighbor_distances.mean(axis=1)
    
    return {
        'mean_neighbor_distance': mean_nn_dist.mean(),
        'std_neighbor_distance': mean_nn_dist.std(),
        'min_neighbor_distance': neighbor_distances[:, 0].min(),
        'max_neighbor_distance': neighbor_distances[:, 0].max(),
        'median_neighbor_distance': np.median(neighbor_distances[:, 0]),
        'neighbor_distances_per_particle': mean_nn_dist,
    }


def compute_radial_distribution_function(positions: np.ndarray, 
                                         max_radius: float = None, 
                                         n_bins: int = 50) -> Tuple[np.ndarray, np.ndarray]:
    """计算径向分布函数 (RDF/g(r))"""
    N = len(positions)
    
    if max_radius is None:
        x_range = positions[:, 0].max() - positions[:, 0].min()
        y_range = positions[:, 1].max() - positions[:, 1].min()
        max_radius = min(x_range, y_range) / 4
    
    # 计算所有粒子对之间的距离
    tree = cKDTree(positions)
    pairs = tree.query_pairs(max_radius, output_type='ndarray')
    
    if len(pairs) == 0:
        return np.linspace(0, max_radius, n_bins), np.zeros(n_bins)
    
    # 计算距离
    distances = np.linalg.norm(positions[pairs[:, 0]] - positions[pairs[:, 1]], axis=1)
    
    # 创建直方图
    hist, bin_edges = np.histogram(distances, bins=n_bins, range=(0, max_radius))
    r = (bin_edges[:-1] + bin_edges[1:]) / 2
    dr = bin_edges[1] - bin_edges[0]
    
    # 归一化
    area = (positions[:, 0].max() - positions[:, 0].min()) * \
           (positions[:, 1].max() - positions[:, 1].min())
    density = N / (area + 1e-10)
    
    # 环形面积
    shell_area = 2 * np.pi * r * dr
    
    # 归一化RDF
    g_r = hist / (shell_area * density * N + 1e-10)
    
    return r, g_r


def compute_voronoi_statistics(positions: np.ndarray, area_percentile_threshold: float = 99.0) -> Dict:
    """
    计算Voronoi图的统计特性
    
    Args:
        positions: 粒子位置
        area_percentile_threshold: 过滤异常大面积的百分位阈值（默认99%）
    """
    try:
        vor = Voronoi(positions)
        
        areas = []
        perimeters = []
        
        for region_idx in vor.point_region:
            region = vor.regions[region_idx]
            
            # 跳过无界或空区域
            if -1 in region or len(region) == 0:
                continue
            
            vertices = vor.vertices[region]
            
            # 计算面积
            x, y = vertices[:, 0], vertices[:, 1]
            area = 0.5 * np.abs(np.dot(x, np.roll(y, 1)) - np.dot(y, np.roll(x, 1)))
            
            # 计算周长
            perimeter = np.sum(np.linalg.norm(vertices - np.roll(vertices, 1, axis=0), axis=1))
            
            areas.append(area)
            perimeters.append(perimeter)
        
        if len(areas) == 0:
            return {
                'mean_voronoi_area': np.nan,
                'std_voronoi_area': np.nan,
                'cv_voronoi_area': np.nan,
                'mean_voronoi_perimeter': np.nan,
                'std_voronoi_perimeter': np.nan
            }
        
        areas = np.array(areas)
        perimeters = np.array(perimeters)
        
        # 过滤异常大的面积（可能是边界效应导致的）
        area_threshold = np.percentile(areas, area_percentile_threshold)
        valid_mask = areas <= area_threshold
        
        if valid_mask.sum() == 0:
            # 如果所有都被过滤了，使用原始数据
            valid_areas = areas
            valid_perimeters = perimeters
        else:
            valid_areas = areas[valid_mask]
            valid_perimeters = perimeters[valid_mask]
        
        return {
            'mean_voronoi_area': valid_areas.mean(),
            'std_voronoi_area': valid_areas.std(),
            'cv_voronoi_area': valid_areas.std() / (valid_areas.mean() + 1e-10),
            'mean_voronoi_perimeter': valid_perimeters.mean(),
            'std_voronoi_perimeter': valid_perimeters.std(),
            'voronoi_areas': valid_areas,
            'voronoi_perimeters': valid_perimeters,
            'n_filtered': len(areas) - valid_mask.sum()
        }
    except Exception as e:
        print(f"Voronoi计算失败: {e}")
        return {
            'mean_voronoi_area': np.nan,
            'std_voronoi_area': np.nan,
            'cv_voronoi_area': np.nan,
            'mean_voronoi_perimeter': np.nan,
            'std_voronoi_perimeter': np.nan
        }


def compute_order_parameter(velocities: np.ndarray) -> float:
    """计算序参数（极化度）"""
    if len(velocities) == 0:
        return 0.0
    
    speeds = np.linalg.norm(velocities, axis=1, keepdims=True)
    speeds = np.where(speeds < 1e-10, 1e-10, speeds)
    
    normalized_velocities = velocities / speeds
    mean_direction = normalized_velocities.mean(axis=0)
    order_param = np.linalg.norm(mean_direction)
    
    return order_param


def compute_comprehensive_metrics(pred_trajectory: np.ndarray, 
                                  gt_trajectory: np.ndarray = None,
                                  history_trajectory: np.ndarray = None,
                                  k_neighbors: int = 6) -> Dict:
    """综合计算所有时间步的指标"""
    T_pred, N, _ = pred_trajectory.shape
    
    # 初始化结果字典
    results = {
        'pred': {
            'density_entropy': [],
            'mean_neighbor_distance': [],
            'std_neighbor_distance': [],
            'mean_voronoi_area': [],
            'std_voronoi_area': [],
            'cv_voronoi_area': [],
            'order_parameter': [],
            'rdf_data': []
        }
    }
    
    if gt_trajectory is not None:
        results['gt'] = {
            'density_entropy': [],
            'mean_neighbor_distance': [],
            'std_neighbor_distance': [],
            'mean_voronoi_area': [],
            'std_voronoi_area': [],
            'cv_voronoi_area': [],
            'order_parameter': [],
            'rdf_data': []
        }
        
        results['diff'] = {
            'density_entropy': [],
            'mean_neighbor_distance': [],
            'mean_voronoi_area': [],
            'order_parameter': []
        }
    
    if history_trajectory is not None:
        results['history'] = {
            'density_entropy': [],
            'mean_neighbor_distance': [],
            'std_neighbor_distance': [],
            'mean_voronoi_area': [],
            'std_voronoi_area': [],
            'cv_voronoi_area': [],
            'order_parameter': [],
            'rdf_data': []
        }
        
        T_hist = history_trajectory.shape[0]
        print(f"  Computing metrics for {T_hist} history frames...")
        
        # 计算历史帧的指标
        for t in range(T_hist):
            hist_pos = history_trajectory[t]
            
            density_metrics = compute_spatial_density_metrics(hist_pos)
            neighbor_metrics = compute_neighbor_statistics(hist_pos, k=k_neighbors)
            voronoi_metrics = compute_voronoi_statistics(hist_pos)
            
            results['history']['density_entropy'].append(density_metrics['density_entropy'])
            results['history']['mean_neighbor_distance'].append(neighbor_metrics['mean_neighbor_distance'])
            results['history']['std_neighbor_distance'].append(neighbor_metrics['std_neighbor_distance'])
            results['history']['mean_voronoi_area'].append(voronoi_metrics['mean_voronoi_area'])
            results['history']['std_voronoi_area'].append(voronoi_metrics['std_voronoi_area'])
            results['history']['cv_voronoi_area'].append(voronoi_metrics['cv_voronoi_area'])
            
            r, g_r = compute_radial_distribution_function(hist_pos)
            results['history']['rdf_data'].append((r, g_r))
            
            if t > 0:
                hist_vel = history_trajectory[t] - history_trajectory[t-1]
                order_hist = compute_order_parameter(hist_vel)
                results['history']['order_parameter'].append(order_hist)
    
    print(f"  Computing metrics for {T_pred} prediction frames...")
    
    # 计算预测帧的指标
    for t in range(T_pred):
        pred_pos = pred_trajectory[t]
        
        density_metrics = compute_spatial_density_metrics(pred_pos)
        neighbor_metrics = compute_neighbor_statistics(pred_pos, k=k_neighbors)
        voronoi_metrics = compute_voronoi_statistics(pred_pos)
        
        results['pred']['density_entropy'].append(density_metrics['density_entropy'])
        results['pred']['mean_neighbor_distance'].append(neighbor_metrics['mean_neighbor_distance'])
        results['pred']['std_neighbor_distance'].append(neighbor_metrics['std_neighbor_distance'])
        results['pred']['mean_voronoi_area'].append(voronoi_metrics['mean_voronoi_area'])
        results['pred']['std_voronoi_area'].append(voronoi_metrics['std_voronoi_area'])
        results['pred']['cv_voronoi_area'].append(voronoi_metrics['cv_voronoi_area'])
        
        r, g_r = compute_radial_distribution_function(pred_pos)
        results['pred']['rdf_data'].append((r, g_r))
        
        if t > 0:
            pred_vel = pred_trajectory[t] - pred_trajectory[t-1]
            order_pred = compute_order_parameter(pred_vel)
            results['pred']['order_parameter'].append(order_pred)
        
        # GT指标
        if gt_trajectory is not None:
            gt_pos = gt_trajectory[t]
            
            gt_density = compute_spatial_density_metrics(gt_pos)
            gt_neighbor = compute_neighbor_statistics(gt_pos, k=k_neighbors)
            gt_voronoi = compute_voronoi_statistics(gt_pos)
            
            results['gt']['density_entropy'].append(gt_density['density_entropy'])
            results['gt']['mean_neighbor_distance'].append(gt_neighbor['mean_neighbor_distance'])
            results['gt']['std_neighbor_distance'].append(gt_neighbor['std_neighbor_distance'])
            results['gt']['mean_voronoi_area'].append(gt_voronoi['mean_voronoi_area'])
            results['gt']['std_voronoi_area'].append(gt_voronoi['std_voronoi_area'])
            results['gt']['cv_voronoi_area'].append(gt_voronoi['cv_voronoi_area'])
            
            r_gt, g_r_gt = compute_radial_distribution_function(gt_pos)
            results['gt']['rdf_data'].append((r_gt, g_r_gt))
            
            if t > 0:
                gt_vel = gt_trajectory[t] - gt_trajectory[t-1]
                order_gt = compute_order_parameter(gt_vel)
                results['gt']['order_parameter'].append(order_gt)
                
                results['diff']['order_parameter'].append(abs(order_pred - order_gt))
            
            results['diff']['density_entropy'].append(
                abs(density_metrics['density_entropy'] - gt_density['density_entropy'])
            )
            results['diff']['mean_neighbor_distance'].append(
                abs(neighbor_metrics['mean_neighbor_distance'] - gt_neighbor['mean_neighbor_distance'])
            )
            results['diff']['mean_voronoi_area'].append(
                abs(voronoi_metrics['mean_voronoi_area'] - gt_voronoi['mean_voronoi_area'])
            )
    
    return results


# ====================== 独立绘图函数 ======================

def plot_density_entropy(metrics: Dict, save_path: str = None):
    """绘制密度熵"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['density_entropy'])
    T_hist = len(metrics['history']['density_entropy']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['density_entropy'], 
               'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['density_entropy'], 
           'b-', linewidth=3, label='Predicted', marker='o', markersize=6, 
           markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['density_entropy'], 
               'r--', linewidth=3, label='Ground Truth', marker='s', 
               markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Density Entropy', fontsize=16, fontweight='bold')
    ax.set_title('Spatial Distribution Entropy', 
                fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Density entropy plot saved to {save_path}")
    
    plt.close()


def plot_mean_neighbor_distance(metrics: Dict, save_path: str = None):
    """绘制平均邻居距离"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['mean_neighbor_distance'])
    T_hist = len(metrics['history']['mean_neighbor_distance']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['mean_neighbor_distance'], 
                'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['mean_neighbor_distance'], 
            'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
            markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['mean_neighbor_distance'], 
                'r--', linewidth=3, label='Ground Truth', marker='s', 
                markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Mean Neighbor Distance', fontsize=16, fontweight='bold')
    ax.set_title('Mean Neighbor Distance', fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Mean neighbor distance plot saved to {save_path}")
    
    plt.close()


def plot_std_neighbor_distance(metrics: Dict, save_path: str = None):
    """绘制邻居距离标准差"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['std_neighbor_distance'])
    T_hist = len(metrics['history']['std_neighbor_distance']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['std_neighbor_distance'], 
                'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['std_neighbor_distance'], 
            'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
            markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['std_neighbor_distance'], 
                'r--', linewidth=3, label='Ground Truth', marker='s', 
                markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Std of Neighbor Distance', fontsize=16, fontweight='bold')
    ax.set_title('Neighbor Distance Variability', fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Std neighbor distance plot saved to {save_path}")
    
    plt.close()


def plot_mean_voronoi_area(metrics: Dict, save_path: str = None):
    """绘制平均Voronoi面积"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['mean_voronoi_area'])
    T_hist = len(metrics['history']['mean_voronoi_area']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['mean_voronoi_area'], 
                'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['mean_voronoi_area'], 
            'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
            markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['mean_voronoi_area'], 
                'r--', linewidth=3, label='Ground Truth', marker='s', 
                markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Mean Voronoi Area', fontsize=16, fontweight='bold')
    ax.set_title('Mean Voronoi Cell Area', fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Mean Voronoi area plot saved to {save_path}")
    
    plt.close()


def plot_std_voronoi_area(metrics: Dict, save_path: str = None):
    """绘制Voronoi面积标准差"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['std_voronoi_area'])
    T_hist = len(metrics['history']['std_voronoi_area']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['std_voronoi_area'], 
                'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['std_voronoi_area'], 
            'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
            markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['std_voronoi_area'], 
                'r--', linewidth=3, label='Ground Truth', marker='s', 
                markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Std of Voronoi Area', fontsize=16, fontweight='bold')
    ax.set_title('Voronoi Area Standard Deviation', fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Std Voronoi area plot saved to {save_path}")
    
    plt.close()


def plot_cv_voronoi_area(metrics: Dict, save_path: str = None):
    """绘制Voronoi面积变异系数"""
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['cv_voronoi_area'])
    T_hist = len(metrics['history']['cv_voronoi_area']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(T_pred)
    
    if has_history:
        ax.plot(hist_steps, metrics['history']['cv_voronoi_area'], 
                'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['cv_voronoi_area'], 
            'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
            markevery=max(1, T_pred//10))
    
    if has_gt:
        ax.plot(pred_steps, metrics['gt']['cv_voronoi_area'], 
                'r--', linewidth=3, label='Ground Truth', marker='s', 
                markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Coefficient of Variation', fontsize=16, fontweight='bold')
    ax.set_title('Voronoi Area Variability (CV)', fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] CV Voronoi area plot saved to {save_path}")
    
    plt.close()


def plot_rdf(metrics: Dict, save_path: str = None):
    """绘制径向分布函数（2×2子图）"""
    has_gt = 'gt' in metrics
    
    if not has_gt or len(metrics['gt']['rdf_data']) == 0:
        print("[WARNING] No GT RDF data available")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))
    axes = axes.flatten()
    
    T_pred = len(metrics['pred']['rdf_data'])
    
    # 选择GT的中间帧
    gt_frame_idx = T_pred // 2
    
    # 随机选择预测的3帧
    np.random.seed(42)
    if T_pred >= 3:
        segment_size = T_pred // 3
        pred_frame_indices = [
            np.random.randint(0, segment_size),
            np.random.randint(segment_size, 2 * segment_size),
            np.random.randint(2 * segment_size, T_pred)
        ]
    else:
        pred_frame_indices = list(range(T_pred))
        while len(pred_frame_indices) < 3:
            pred_frame_indices.append(pred_frame_indices[-1])
    
    # 子图1：GT
    r_gt, g_r_gt = metrics['gt']['rdf_data'][gt_frame_idx]
    axes[0].plot(r_gt, g_r_gt, 'r-', linewidth=3, alpha=0.9, label='Ground Truth')
    axes[0].axhline(y=1, color='k', linestyle=':', linewidth=2, alpha=0.5)
    axes[0].set_xlabel('Distance r', fontsize=14, fontweight='bold')
    axes[0].set_ylabel('g(r)', fontsize=14, fontweight='bold')
    axes[0].set_title(f'Ground Truth (t={gt_frame_idx})', fontsize=16, fontweight='bold')
    axes[0].legend(fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].tick_params(labelsize=12)
    
    # 子图2-4：预测的3帧
    for i, frame_idx in enumerate(pred_frame_indices):
        ax = axes[i + 1]
        r_pred, g_r_pred = metrics['pred']['rdf_data'][frame_idx]
        ax.plot(r_pred, g_r_pred, 'b-', linewidth=3, alpha=0.9, label='Predicted')
        ax.axhline(y=1, color='k', linestyle=':', linewidth=2, alpha=0.5)
        ax.set_xlabel('Distance r', fontsize=14, fontweight='bold')
        ax.set_ylabel('g(r)', fontsize=14, fontweight='bold')
        ax.set_title(f'Predicted (t={frame_idx})', fontsize=16, fontweight='bold')
        ax.legend(fontsize=12)
        ax.grid(True, alpha=0.3)
        ax.tick_params(labelsize=12)
    
    plt.suptitle('Radial Distribution Function', fontsize=18, fontweight='bold', y=0.995)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] RDF plot saved to {save_path}")
    
    plt.close()


def plot_order_parameter(metrics: Dict, save_path: str = None):
    """绘制序参数"""
    if len(metrics['pred']['order_parameter']) == 0:
        print("[WARNING] No order parameter data available")
        return
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    T_pred = len(metrics['pred']['order_parameter'])
    T_hist = len(metrics['history']['order_parameter']) if has_history else 0
    
    hist_steps = np.arange(-T_hist, 0) if has_history else np.array([])
    pred_steps = np.arange(1, T_pred + 1)
    
    if has_history and T_hist > 0:
        ax.plot(hist_steps, metrics['history']['order_parameter'], 
               'gray', linestyle='--', linewidth=3, alpha=0.7, label='History')
    
    ax.plot(pred_steps, metrics['pred']['order_parameter'], 
           'b-', linewidth=3, label='Predicted', marker='o', markersize=6,
           markevery=max(1, T_pred//10))
    
    if has_gt and len(metrics['gt']['order_parameter']) > 0:
        ax.plot(pred_steps, metrics['gt']['order_parameter'], 
               'r--', linewidth=3, label='Ground Truth', marker='s', 
               markersize=6, markevery=max(1, T_pred//10))
    
    ax.axvline(x=0, color='black', linestyle=':', linewidth=2.5, alpha=0.6)
    
    ax.set_xlabel('Time Step', fontsize=16, fontweight='bold')
    ax.set_ylabel('Order Parameter (ϕ)', fontsize=16, fontweight='bold')
    ax.set_title('Collective Motion Order Parameter', 
                fontsize=18, fontweight='bold', pad=15)
    ax.legend(fontsize=14, loc='best', framealpha=0.9)
    ax.grid(True, alpha=0.3, linewidth=1)
    ax.tick_params(labelsize=14)
    ax.set_ylim([0, 1])
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"[SUCCESS] Order parameter plot saved to {save_path}")
    
    plt.close()


def print_metrics_summary(metrics: Dict, batch_idx: int = 0):
    """打印指标的统计摘要"""
    has_gt = 'gt' in metrics
    has_history = 'history' in metrics
    
    print(f"\n{'='*80}")
    print(f"Batch {batch_idx} - Structural Metrics Summary")
    print(f"{'='*80}")
    
    metric_keys = ['density_entropy', 'mean_neighbor_distance', 'std_neighbor_distance',
                   'mean_voronoi_area', 'std_voronoi_area', 'cv_voronoi_area']
    
    if has_history:
        print(f"\n{'History Frames':^80}")
        print(f"{'-'*80}")
        print(f"{'Metric':<35} {'Mean':<15} {'Std':<15} {'Last Frame':<15}")
        print(f"{'-'*80}")
        
        for key in metric_keys:
            if key in metrics['history']:
                values = np.array(metrics['history'][key])
                valid_values = values[~np.isnan(values)]
                if len(valid_values) > 0:
                    print(f"{key:<35} {valid_values.mean():<15.6f} "
                          f"{valid_values.std():<15.6f} {valid_values[-1]:<15.6f}")
        
        if len(metrics['history']['order_parameter']) > 0:
            order_vals = np.array(metrics['history']['order_parameter'])
            print(f"{'order_parameter':<35} {order_vals.mean():<15.6f} "
                  f"{order_vals.std():<15.6f} {order_vals[-1]:<15.6f}")
    
    print(f"\n{'Predicted Trajectory':^80}")
    print(f"{'-'*80}")
    print(f"{'Metric':<35} {'Mean':<15} {'Std':<15} {'First':<15} {'Last':<15}")
    print(f"{'-'*80}")
    
    for key in metric_keys:
        values = np.array(metrics['pred'][key])
        valid_values = values[~np.isnan(values)]
        if len(valid_values) > 0:
            print(f"{key:<35} {valid_values.mean():<15.6f} "
                  f"{valid_values.std():<15.6f} {valid_values[0]:<15.6f} {valid_values[-1]:<15.6f}")
    
    if len(metrics['pred']['order_parameter']) > 0:
        order_vals = np.array(metrics['pred']['order_parameter'])
        print(f"{'order_parameter':<35} {order_vals.mean():<15.6f} "
              f"{order_vals.std():<15.6f} {order_vals[0]:<15.6f} {order_vals[-1]:<15.6f}")
    
    if has_gt:
        print(f"\n{'Ground Truth Trajectory':^80}")
        print(f"{'-'*80}")
        print(f"{'Metric':<35} {'Mean':<15} {'Std':<15} {'First':<15} {'Last':<15}")
        print(f"{'-'*80}")
        
        for key in metric_keys:
            values = np.array(metrics['gt'][key])
            valid_values = values[~np.isnan(values)]
            if len(valid_values) > 0:
                print(f"{key:<35} {valid_values.mean():<15.6f} "
                      f"{valid_values.std():<15.6f} {valid_values[0]:<15.6f} {valid_values[-1]:<15.6f}")
        
        if len(metrics['gt']['order_parameter']) > 0:
            order_vals = np.array(metrics['gt']['order_parameter'])
            print(f"{'order_parameter':<35} {order_vals.mean():<15.6f} "
                  f"{order_vals.std():<15.6f} {order_vals[0]:<15.6f} {order_vals[-1]:<15.6f}")
        
        print(f"\n{'Prediction Error (Structural Metrics)':^80}")
        print(f"{'-'*80}")
        print(f"{'Metric':<35} {'Mean Abs Diff':<20} {'Max Abs Diff':<20}")
        print(f"{'-'*80}")
        
        for key in ['density_entropy', 'mean_neighbor_distance', 'mean_voronoi_area']:
            if key in metrics['diff']:
                diff_values = np.array(metrics['diff'][key])
                valid_diffs = diff_values[~np.isnan(diff_values)]
                if len(valid_diffs) > 0:
                    print(f"{key:<35} {valid_diffs.mean():<20.6f} {valid_diffs.max():<20.6f}")
        
        if 'order_parameter' in metrics['diff'] and len(metrics['diff']['order_parameter']) > 0:
            order_diff = np.array(metrics['diff']['order_parameter'])
            print(f"{'order_parameter':<35} {order_diff.mean():<20.6f} {order_diff.max():<20.6f}")
    
    print(f"{'='*80}\n")


# ====================== 边构建函数 ======================

def build_edges_from_geometry(pos, original_edge_attr=None, method='knn', k=6, 
                               radius=None, self_loop=False, device='cpu'):
    """基于节点几何位置构建邻接关系"""
    if isinstance(pos, torch.Tensor):
        pos_np = pos.cpu().detach().numpy()
    else:
        pos_np = np.array(pos)
    
    if original_edge_attr is not None:
        if isinstance(original_edge_attr, torch.Tensor):
            edge_attr_dim = original_edge_attr.shape[1] if original_edge_attr.dim() > 1 else 1
        else:
            edge_attr_dim = original_edge_attr.shape[1] if len(original_edge_attr.shape) > 1 else 1
    else:
        edge_attr_dim = 3
    
    N = pos_np.shape[0]
    
    if method == 'knn':
        edge_index, distances = _build_knn_edges(pos_np, k, self_loop)
    elif method == 'radius':
        if radius is None:
            radius = np.mean(np.linalg.norm(pos_np - pos_np.mean(axis=0), axis=1)) * 0.3
        edge_index, distances = _build_radius_edges(pos_np, radius, self_loop)
    elif method == 'delaunay':
        edge_index, distances = _build_delaunay_edges(pos_np, self_loop)
    else:
        raise ValueError(f"未知的构建方法: {method}")
    
    if edge_index.shape[1] > 0:
        edge_attr = _generate_edge_attributes(pos_np, edge_index, distances, edge_attr_dim)
    else:
        edge_attr = np.zeros((0, edge_attr_dim), dtype=np.float32)
    
    edge_index = torch.tensor(edge_index, dtype=torch.long, device=device)
    edge_attr = torch.tensor(edge_attr, dtype=torch.float32, device=device)
    
    return edge_index, edge_attr


def _build_knn_edges(pos_np, k, self_loop=False):
    """KNN方法构建边"""
    N = pos_np.shape[0]
    k = min(k, N - 1)
    
    tree = cKDTree(pos_np)
    distances, indices = tree.query(pos_np, k=k + 1)
    
    edge_index_list = []
    distances_list = []
    
    for i in range(N):
        neighbors = indices[i, 1:]
        dists = distances[i, 1:]
        
        for j, neighbor in enumerate(neighbors):
            if neighbor < N:
                edge_index_list.append([i, neighbor])
                distances_list.append(dists[j])
    
    edge_index = np.array(edge_index_list).T if edge_index_list else np.zeros((2, 0), dtype=np.long)
    distances = np.array(distances_list)
    
    return edge_index, distances


def _build_radius_edges(pos_np, radius, self_loop=False):
    """半径方法构建边"""
    tree = cKDTree(pos_np)
    pairs = tree.query_pairs(radius, output_type='ndarray')
    
    if len(pairs) == 0:
        return np.zeros((2, 0), dtype=np.long), np.array([])
    
    edge_index_list = []
    distances_list = []
    
    for i, j in pairs:
        dist = np.linalg.norm(pos_np[j] - pos_np[i])
        edge_index_list.append([i, j])
        edge_index_list.append([j, i])
        distances_list.append(dist)
        distances_list.append(dist)
    
    edge_index = np.array(edge_index_list).T
    distances = np.array(distances_list)
    
    return edge_index, distances


def _build_delaunay_edges(pos_np, self_loop=False):
    """Delaunay三角剖分构建边"""
    N = pos_np.shape[0]
    
    if N < 3:
        return _build_knn_edges(pos_np, k=min(2, N-1), self_loop=self_loop)
    
    try:
        tri = Delaunay(pos_np)
    except Exception:
        return _build_knn_edges(pos_np, k=min(6, N-1), self_loop=self_loop)
    
    edge_set = set()
    for simplex in tri.simplices:
        for i in range(3):
            for j in range(i + 1, 3):
                edge = tuple(sorted([simplex[i], simplex[j]]))
                edge_set.add(edge)
    
    edge_index_list = []
    distances_list = []
    
    for i, j in edge_set:
        for src, tgt in [(i, j), (j, i)]:
            dist = np.linalg.norm(pos_np[tgt] - pos_np[src])
            edge_index_list.append([src, tgt])
            distances_list.append(dist)
    
    edge_index = np.array(edge_index_list).T if edge_index_list else np.zeros((2, 0), dtype=np.long)
    distances = np.array(distances_list)
    
    return edge_index, distances


def _generate_edge_attributes(pos_np, edge_index, distances, dim):
    """生成边属性"""
    E = edge_index.shape[1]
    edge_attr = np.zeros((E, dim), dtype=np.float32)
    
    if E == 0:
        return edge_attr
    
    edge_attr[:, 0] = distances
    
    if dim >= 2:
        directions = pos_np[edge_index[1]] - pos_np[edge_index[0]]
        edge_attr[:, 1] = directions[:, 0]
    
    if dim >= 3:
        edge_attr[:, 2] = directions[:, 1]
    
    if dim >= 4:
        edge_attr[:, 3] = distances ** 2
    
    if dim >= 5:
        edge_attr[:, 4] = np.linalg.norm(directions, axis=1)
    
    if dim > 5:
        edge_attr[:, 5:] = 0.0
    
    return edge_attr


# ====================== 速度特征计算 ======================

def compute_velocity_features(graph_x, sequence_len):
    """将绝对位置特征转换为速度特征"""
    time_feature = graph_x[:, -1:]
    pos_features = graph_x[:, :-1]
    
    num_nodes = pos_features.size(0)
    pos_seq = pos_features.view(num_nodes, sequence_len, 2)
    
    velocities = pos_seq[:, 1:, :] - pos_seq[:, :-1, :]
    
    zeros = torch.zeros(num_nodes, 1, 2, device=graph_x.device)
    velocities_padded = torch.cat([zeros, velocities], dim=1)
    
    velocities_flat = velocities_padded.view(num_nodes, -1)
    new_features = torch.cat([velocities_flat, time_feature], dim=1)
    
    return new_features


# ====================== 推理函数 ======================

def inference(num_rollout_steps: int = 10, edge_method='knn', edge_k=6, 
              start_frame_idx: int = 0, correction_interval: int = 3):
    """执行推理并返回轨迹"""
    all_predictions = []
    all_gt_trajectories = []  
    all_initial_positions = []
    all_histories = []
    all_correction_flags = []

    num_history_features = sequence_length * 2 
    
    # 收集所有帧
    all_frame_batches_raw = []
    
    for batch_idx, batch_graph in enumerate(test_loader):
        try:
            batch_graph = transformer(batch_graph)
        except Exception:
            batch_graph = transformer(batch_graph.to(device)).to('cpu')
        
        batch_graph = batch_graph.to(device)
        all_frame_batches_raw.append(batch_graph)
    
    if start_frame_idx >= len(all_frame_batches_raw):
        raise ValueError(f"start_frame_idx ({start_frame_idx}) 超出范围")
        
    all_frame_batches = all_frame_batches_raw[start_frame_idx:]
    num_frames = len(all_frame_batches)
    first_batch = all_frame_batches[0]
    num_graphs_per_batch = len(first_batch.ptr) - 1
    
    effective_rollout_steps = min(num_rollout_steps, num_frames - 1)
    
    print(f"执行 Rollout 预测（{effective_rollout_steps} 步）...")
    
    for graph_idx in range(num_graphs_per_batch):
        initial_frame_batch = all_frame_batches[0]
        start_node = initial_frame_batch.ptr[graph_idx].item()
        end_node = initial_frame_batch.ptr[graph_idx + 1].item()
        node_indices = torch.arange(start_node, end_node)
        
        initial_positions = initial_frame_batch.pos[node_indices].clone()
        all_initial_positions.append(initial_positions.cpu().detach().numpy())

        current_x = initial_frame_batch.x[node_indices].clone()
        current_pos = initial_positions.clone()
        
        # 提取历史轨迹
        history_flat = current_x[:, :num_history_features]
        history_unflattened = history_flat.view(history_flat.shape[0], sequence_length, 2)
        history_trajectory = history_unflattened.permute(1, 0, 2).cpu().detach().numpy()
        current_pos_np = initial_positions.cpu().detach().numpy()
        current_pos_expanded = np.expand_dims(current_pos_np, axis=0)
        full_history = np.concatenate([history_trajectory, current_pos_expanded], axis=0)
        all_histories.append(full_history)

        sample_predictions = [initial_positions.cpu().detach().numpy()]
        sample_correction_flags = [False]
        sample_gt_trajectory = [initial_positions.cpu().detach().numpy()]
        
        # 预先获取GT数据
        gt_data_all = []
        gt_positions_all = [initial_positions.cpu().detach().numpy()]
        
        for frame_idx in range(effective_rollout_steps + 1):
            if frame_idx < num_frames:
                frame_batch = all_frame_batches[frame_idx]
                s_n = frame_batch.ptr[graph_idx].item()
                e_n = frame_batch.ptr[graph_idx + 1].item()
                indices = torch.arange(s_n, e_n)
                
                pos = frame_batch.pos[indices]
                x_feat = frame_batch.x[indices]
                
                if frame_idx > 0:
                    sample_gt_trajectory.append(pos.cpu().detach().numpy())
                    gt_positions_all.append(pos.cpu().detach().numpy())
                
                gt_data_all.append({'pos': pos, 'x': x_feat})

        original_edge_attr = initial_frame_batch.edge_attr
        
        current_edge_index, current_edge_attr = build_edges_from_geometry(
            pos=current_pos,
            original_edge_attr=original_edge_attr,
            method=edge_method,
            k=edge_k,
            self_loop=False,
            device=device
        )

        steps_since_correction = 0 

        # Rollout循环
        for step in range(effective_rollout_steps):
            use_correction = (steps_since_correction >= correction_interval) and \
                           (step + 1 < len(gt_data_all))
            
            if use_correction:
                gt_step_data = gt_data_all[step + 1]
                predicted_positions = gt_step_data['pos'].clone() 
                current_x = gt_step_data['x'].clone()
                
                sample_predictions.append(predicted_positions.cpu().detach().numpy())
                sample_correction_flags.append(True)
                
                current_pos = predicted_positions
                steps_since_correction = 0
                
            else:
                with torch.no_grad():
                    temp_graph = Data(
                        x=compute_velocity_features(current_x, sequence_length),
                        pos=current_pos,
                        edge_index=current_edge_index,
                        edge_attr=current_edge_attr
                    )
                    
                    pred_disp_norm = model(temp_graph)

                    if isinstance(pred_disp_norm, (tuple, list)):
                        pred_disp_norm = pred_disp_norm[0]
                    if pred_disp_norm.shape != current_pos.shape:
                        pred_disp_norm = pred_disp_norm.view_as(current_pos)

                    pred_disp = simulator._output_normalizer.inverse(pred_disp_norm)
                
                predicted_positions = current_pos + pred_disp
                sample_predictions.append(predicted_positions.cpu().detach().numpy())
                sample_correction_flags.append(False)

                history = current_x[:, :num_history_features]
                history_shifted = history[:, 2:]
                new_frame = predicted_positions
                updated_history = torch.cat([history_shifted, new_frame], dim=1)

                time_col = current_x[:, num_history_features:num_history_features+1]
                new_time_col = time_col + time_step

                current_x = torch.cat([updated_history, new_time_col], dim=1)
                current_pos = predicted_positions
                
                steps_since_correction += 1

            current_edge_index, current_edge_attr = build_edges_from_geometry(
                pos=current_pos,
                original_edge_attr=original_edge_attr,
                method=edge_method,
                k=edge_k,
                self_loop=False,
                device=device
            )
        
        all_predictions.append(np.stack(sample_predictions, axis=0))
        all_gt_trajectories.append(np.stack(sample_gt_trajectory, axis=0))
        all_correction_flags.append(sample_correction_flags)
    
    return all_initial_positions, all_predictions, all_gt_trajectories, all_histories, all_correction_flags


# ====================== 主程序 ======================

if __name__ == "__main__":
    # 创建模拟器并加载模型
    simulator = Simulator(
        message_passing_num=message_passing_num,
        node_input_size=node_input_size,
        edge_input_size=edge_input_size,
        device=device,
        hidden_size=hidden_size,
        model_dir=load_dir
    )

    try:
        loaded = simulator.load_checkpoint()
    except TypeError:
        loaded = simulator.load_checkpoint(load_dir)

    if hasattr(simulator, 'model') and simulator.model is not None:
        model = simulator.model
    else:
        model = loaded
    
    if model is None:
        raise RuntimeError("无法加载模型")
    
    model.to(device)
    model.eval()

    # 数据转换
    transformer = T.Compose([
        T.Cartesian(norm=False),
        T.Distance(norm=False)
    ])

    # 加载数据集
    dataset_cell = CellMotionRollout(
        dataset_dir=dataset_dir, 
        sequence_length=sequence_length, 
        split='test_e0n300'
    )
    test_loader = DataLoader(dataset=dataset_cell, batch_size=batch_size, num_workers=0)

    # 执行推理
    print("\n" + "="*80)
    print("开始推理...")
    print("="*80)
    
    initial_positions, predictions, ground_truth, history_frames, correction_flags = inference(
        num_rollout_steps=max_rollout_steps, 
        edge_method='delaunay',
        start_frame_idx=start_frame_idx,
        correction_interval=correction_interval  
    )
    
    # 打印校正统计
    for i, flags in enumerate(correction_flags):
        num_pred = sum(not f for f in flags)
        num_corr = sum(flags)
        print(f"Batch {i}: 预测帧数={num_pred}, 校正帧数={num_corr}")

    # 裁剪GT轨迹
    predictions_frames = predictions[0].shape[0]  
    ground_truth_cropped = [gt[:predictions_frames] for gt in ground_truth]

    # ==================== 计算综合评估指标 ====================
    print("\n" + "="*80)
    print("计算综合结构指标...")
    print("="*80)
    
    all_metrics = []
    
    # 创建保存目录
    save_dir = 'metrics_plots'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    for batch_idx in range(min(3, len(predictions))):
        print(f"\n处理 Batch {batch_idx}...")
        
        pred_traj = predictions[batch_idx]
        gt_traj = ground_truth_cropped[batch_idx] if ground_truth_cropped else None
        hist_traj = history_frames[batch_idx] if history_frames else None
        
        # 计算所有指标
        metrics = compute_comprehensive_metrics(
            pred_trajectory=pred_traj,
            gt_trajectory=gt_traj,
            history_trajectory=hist_traj,
            k_neighbors=6
        )
        
        all_metrics.append(metrics)
        
        # 打印摘要
        print_metrics_summary(metrics, batch_idx=batch_idx)
        
        # ========== 分别保存8个指标图 ==========
        print(f"\n生成 Batch {batch_idx} 的指标图...")
        
        plot_density_entropy(metrics, save_path=f'{save_dir}/batch{batch_idx}_density_entropy.png')
        plot_mean_neighbor_distance(metrics, save_path=f'{save_dir}/batch{batch_idx}_mean_neighbor_distance.png')
        plot_std_neighbor_distance(metrics, save_path=f'{save_dir}/batch{batch_idx}_std_neighbor_distance.png')
        plot_mean_voronoi_area(metrics, save_path=f'{save_dir}/batch{batch_idx}_mean_voronoi_area.png')
        plot_std_voronoi_area(metrics, save_path=f'{save_dir}/batch{batch_idx}_std_voronoi_area.png')
        plot_cv_voronoi_area(metrics, save_path=f'{save_dir}/batch{batch_idx}_cv_voronoi_area.png')
        plot_rdf(metrics, save_path=f'{save_dir}/batch{batch_idx}_rdf.png')
        plot_order_parameter(metrics, save_path=f'{save_dir}/batch{batch_idx}_order_parameter.png')
    
    # ==================== 保存指标数据 ====================
    metrics_save_path = 'evaluation_metrics.pkl'
    with open(metrics_save_path, 'wb') as f:
        pickle.dump(all_metrics, f)
    print(f"\n[SUCCESS] 所有指标已保存到 {metrics_save_path}")
    
    print("\n" + "="*80)
    print("所有任务完成！")
    print(f"指标图保存在: {save_dir}/")
    print(f"共生成 {len(all_metrics) * 8} 张独立图片")
    print("="*80)
```
